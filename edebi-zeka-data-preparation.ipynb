```python
import pandas as pd
import numpy as np
from nltk.corpus import stopwords
```

```python
stopwords = stopwords.words("turkish")
edebi_zeka_df_age = pd.DataFrame(columns=["Dize", "Dönem"])
edebi_zeka_df_century = pd.DataFrame(columns=["Dize", "Yüzyıl"])
edebi_zeka_df_poet = pd.DataFrame(columns=["Dize", "Şair"])
```

```python
def data_prep_age(file_name, age):
    file_name += ".txt"
    f = open(file_name, "r", encoding="utf-8")
    veri = f.read()
    veri = veri.lower().split("\n")
    
    for i in veri:
        for x in i:
            if x in stopwords:
                i.replace(x, "")
    try:
        for i in veri:
            olcum_ind = veri.index(i)
            if "...." in veri[olcum_ind+1]:
                veri.remove(i)
                
    except:
        pass
    
    for i in veri:
        if "...." in i:
            veri.remove(i)    
    for i in veri:
        if i == "":
            veri.remove(i)
            
    for i in veri:
        if i == " ":
            veri.remove(i)
            
    veri_df = pd.DataFrame(veri, columns=["Dize"])
    veri_df["Dize"] = veri_df["Dize"].str.replace("[^\w\s]", "")
    veri_df["Dize"] = veri_df["Dize"].str.replace("\d", "")
    veri_df.replace("", np.nan, inplace=True)
    veri_df.dropna(subset=["Dize"], inplace=True)        
    veri_df["Dönem"] = age
    return veri_df

def data_prep_poet(file_name, poet):
    file_name += ".txt"
    f = open(file_name, "r", encoding="utf-8")
    veri = f.read()
    veri = veri.lower().split("\n")
    
    for i in veri:
        for x in i:
            if x in stopwords:
                i.replace(x, "")
    try:
        for i in veri:
            olcum_ind = veri.index(i)
            if "...." in veri[olcum_ind+1]:
                veri.remove(i)
                
    except:
        pass
    
    for i in veri:
        if "...." in i:
            veri.remove(i)    
    for i in veri:
        if i == "":
            veri.remove(i)
            
    for i in veri:
        if i == " ":
            veri.remove(i)
            
    veri_df = pd.DataFrame(veri, columns=["Dize"])
    veri_df["Dize"] = veri_df["Dize"].str.replace("[^\w\s]", "")
    veri_df["Dize"] = veri_df["Dize"].str.replace("\d", "")
    veri_df.replace("", np.nan, inplace=True)
    veri_df.dropna(subset=["Dize"], inplace=True)        
    veri_df["Şair"] = poet
    return veri_df

def data_prep_century(file_name, century):
    file_name += ".txt"
    f = open(file_name, "r", encoding="utf-8")
    veri = f.read()
    veri = veri.lower().split("\n")
    
    for i in veri:
        for x in i:
            if x in stopwords:
                i.replace(x, "")
    try:
        for i in veri:
            olcum_ind = veri.index(i)
            if "...." in veri[olcum_ind+1]:
                veri.remove(i)
                
    except:
        pass
    
    for i in veri:
        if "...." in i:
            veri.remove(i)    
    for i in veri:
        if i == "":
            veri.remove(i)
            
    for i in veri:
        if i == " ":
            veri.remove(i)
            
    veri_df = pd.DataFrame(veri, columns=["Dize"])
    veri_df["Dize"] = veri_df["Dize"].str.replace("[^\w\s]", "")
    veri_df["Dize"] = veri_df["Dize"].str.replace("\d", "")
    veri_df.replace("", np.nan, inplace=True)
    veri_df.dropna(subset=["Dize"], inplace=True)        
    veri_df["Yüzyıl"] = century
    return veri_df
```

```python
def dataframe_check(df):
    rec_bin = []
    for i in range(len(df)):
        if len(df["Dize"][i].split()) < 3:
            rec_bin.append(i)
            
    cleaned_df = df.drop(rec_bin)
    return cleaned_df
```

```python
def excel_transfer_age(dataframe):
    dataframe.to_excel("edebi-zeka-donem-veri-seti.xlsx", index=False)
    
def excel_transfer_poet(dataframe):
    dataframe.to_excel("edebi-zeka-sair-veri-seti.xlsx", index=False)
    
def excel_transfer_century(dataframe):
    dataframe.to_excel("edebi-zeka-yuzyıl-veri-seti.xlsx", index=False)
```
